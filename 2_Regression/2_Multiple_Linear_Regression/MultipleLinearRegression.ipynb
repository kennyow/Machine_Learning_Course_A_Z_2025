# Multiple Linear Regression

5 Methods of building models:
- All-in
- Backward Elimination
- Forward elimination
- Bidirectional Elimination
- Score Comparison

## All in
- Prior knowledge or
- You have to or
- Preparing for Backward Elimination

# Backward Elimination
- S1: Select a significance level to stay in the model (e.g. SL = 0.05)
- S2: Fit ethe full model with all possible predictors
- S3: Consider the predictor with the highest P value. If P > SL, go to S4, otherwise go to FIN
- S4: Remove the predictor
- S5: Fit model without this variable
- 

# Foward Selection
- S1: Select a significance level to stay in the model (e.g. SL = 0.05)
- S2: Fit all simple regression models y ~xn. Select the one with the lowest P Value
- S3: Keep this variable and fit all possiblemodels with one extra predictor added to the one(s) you already have
- S4: Consider the predictor with the lowest P value. If P < SL, go to S3, otherwise go to FIN
- 

# Bidirectional Elimination
- S1: Select a significance level to stay in the model (e.g. SLENTER = 0.05, SLSTAY = 0.05)
- S2: Perform the next step of Foward Selection (new variables must have: P < SLENTER to enter)
- S3: Perform ALL steps of Backward Elimination( old variables must have P < SLSTAY to stay)
- S4: No new variables can enter and no old variables can exit
- 

# All Possible Models
- S1: Select a criterion of goodness of fit(e.g. Akaike criterion)
- S2: Construct ALL possible regression models: 2^N-1 total combinations)
- S3: Select the one with the best criterion


## Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Import Dataset

```python
dataset = pd.read_csv('50_Startups.csv')
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
dataset.head(5)
```

```python
print(X)
```

```python
print(y) #Profit
```

## Encoding Categorical Data

### Encoding Independent Variable

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
import numpy as np

ct = ColumnTransformer(
    transformers=[('encoder', OneHotEncoder(), [3])],
    remainder='passthrough'
)

X = np.array(ct.fit_transform(X))
```

```python
print(X)
```

### Encoding Dependent Variable

```python
print(X) #R&D Spend, Administration, Marketing Spend, State
```

## Splitting the dataset into Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Training the Multiple Linear Regression Model on the Training Set 

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor. fit(X_train, y_train)
```

## Predicting the Test set results

```python
y_pred = regressor.predict(X_test)#prediction
```

```python
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), 1)) #reshape to concatenate. Y_test contains the real profit
```
